## Шаг 2: Предобработка данных


```python
# 2.1. Обработка аномалий в 'ceiling_height'.  Выявляем и корректируем ошибки ввода, где высота потолков указана некорректно
#  (слишком большие значения, скорее всего, из-за неправильно поставленной запятой).  Применяем корректировку,  разделив
#  значения, превышающие 24 метра, на 10.

data.loc[data['ceiling_height'] >= 24, 'ceiling_height'] = data.loc[data['ceiling_height'] >= 24, 'ceiling_height'] / 10

# 2.2. Заполнение пропусков в 'ceiling_height' медианным значением. Используем медиану, т.к. она менее чувствительна к выбросам.
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())

# 2.3. Обработка пропусков в 'is_apartment'.  Заполняем пропуски False, предполагая, что пропуски возникли из-за того,
# что квартира не для найма.
data['is_apartment'] = data['is_apartment'].fillna(False)

# Преобразуем 'is_apartment' в тип `bool`, т.к. пропуски обработаны.
data['is_apartment'] = data['is_apartment'].astype('bool')

# 2.4. Обработка пропусков в 'balcony'.  Заполняем пропуски 0, предполагая,
# что количество балконов не указано из-за их отсутствия.
data['balcony'] = data['balcony'].fillna(0)

# Преобразуем 'balcony' в тип `int`.
data['balcony'] = data['balcony'].astype('int')

# 2.5. Заполнение пропусков в 'living_area' и 'kitchen_area' медианными значениями.  Используем медиану.
data['living_area'] = data['living_area'].fillna(data['living_area'].median())
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())

# 2.6. Удаление строк с пропусками в 'locality_name' и 'floors_total'.  Эти пропуски нельзя достоверно заполнить.
#  Удаляем, т.к. их доля незначительна (< 0.3%).
cleaned_data = data.dropna(subset=['locality_name', 'floors_total']).copy()

# 2.7. Преобразование типа данных floors_total на int.
cleaned_data['floors_total'] = cleaned_data['floors_total'].astype('int').copy()

# 2.8. Обработка неявных дубликатов в 'locality_name'. Приведение названий населенных пунктов к единообразному виду.
cleaned_data['locality_name'] = (
    cleaned_data['locality_name']
    .str.replace('ё', 'е')
    .replace(['поселок городского типа Мурино', 'поселок Мурино'], 'Мурино', regex=True)
)

# 2.9. Создание новых столбцов.  square_metr - цена одного квадратного метра.
cleaned_data['square_metr'] = round(cleaned_data['last_price'] / cleaned_data['total_area'])

#  Создание столбцов с информацией о дате, дне недели, месяце и годе публикации объявления.
cleaned_data['weekday'] = pd.to_datetime(cleaned_data['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.weekday
cleaned_data['month'] = pd.to_datetime(cleaned_data['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.month
cleaned_data['year'] = pd.to_datetime(cleaned_data['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.year

# Создание столбца с информацией о типе этажа квартиры (первый, последний, другой).
def type_of_floor(data):
    if data['floor'] == 1:
        return 'первый'
    elif data['floor'] == data['floors_total']:
        return 'последний'
    else:
        return 'другой'
cleaned_data['type_of_floor'] = cleaned_data.apply(type_of_floor, axis=1)

# Создание столбца с расстоянием до центра города в километрах (округление до целого числа).
cleaned_data['cityCenters_nearest'] = round(cleaned_data['cityCenters_nearest'] * 0.001)

```
